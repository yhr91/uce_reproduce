```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
```

```python
import pandas as pd
```

```python
distance_vs_loss = pd.DataFrame()
distance_vs_loss["Evolutionary Distance"] = [0, 1, 1, 2, 2, 3, 4, 4, 5, 6]
distance_vs_loss["Loss"] = [0.383, 0.446, 0.458, 0.388, 0.500, 0.461, 0.511, 0.587, 0.422, 0.713]
```

```python
ax = sns.lineplot(data=distance_vs_loss, x="Evolutionary Distance", y="Loss", err_style="bars", errorbar=("pi", 100))
ax.set_ylim((0, 1))
ax.set_title("Loss versus Evolutionary Distance");
```
